```python
import re

def systemize_groupSizes(gsString):
    gsString = gsString.lower().replace(" ","") 
    if gsString.startswith("avg:"):
        hasAvg = True
    gsList = re.split('\,|\;', gsString)
    parsedList = []
    for f in gsList:
        f = f.replace("avg:","").replace("avg.","").replace("avg","")  
        f = re.sub("[^0-9\.\-]", "", f)
        try:
            if "-" in f: # range value, use just average
                fList = f.split("-")
                fList = [float(i) for i in fList]
                f = np.round(np.mean(fList))

            f = np.round(float(f))
            if f >0:
                #print(f)
                parsedList.append(f)
        except:
            print("error in "+gsString, gsList, f)
    return parsedList            
```

```python
ignored_baselines = {"-", "proprietary", "Unclear / not specified", "\"brute-force\"",
                     "individualrsongroupdata", "item-based", "somescalarizationbasedgrs", "unknown", "none",
                    "random", "item.based", "clustering"}
broken_baselines = set()
good_baselines = dict()

def get_baselines(row_baselines):
    baselines = dict()
    pattern = r'\([^()]*\)'
    # Substitute the matched patterns with an empty string
    result = re.sub(pattern, '', row_baselines)
    # Remove any extra commas and whitespace left after the substitution
    result = re.sub(r'\s*,\s*', ',', result).strip(',')
    
    for b in result.replace(";",",").split(","):
        b = b.strip()
        if b in ignored_baselines or not b:
            continue
        tmp = baselines_duplicated[baselines_duplicated.Name.str.lower() == b.lower()]
        #assert len(tmp) <= 1, f"tmp={tmp.Name.values}"
        if len(tmp) == 0:
            pass#print(f"Could not find baseline: '{b}'")
        else:
            the_url = tmp.iloc[0].Paper
            for t in range(len(tmp)):
                url = tmp.iloc[t].Paper
                if (len(raw_data[raw_data.Link == url])) > 0:
                    the_url = url
            #good_baselines[b] = the_url
            baselines[b] = the_url
    return baselines

for _, row in raw_data.iterrows():
    good_baselines.update(get_baselines(row.Baselines))
```

```python
#for each of reported values, get its best fit; then report on the average fit for list 1 
# (i.e., to what extent are sizes used in list1 reflected in list2)
def groupSizeSimilarity(list1, list2):
    # intersection of two sets
    dst = []
    for i in list1:
        dst_i = []
        for j in list2:
            d = abs(i-j) / np.max([i,j])
            dst_i.append(d)
        if len(dst_i) >0:
            dst.append(np.min(dst_i))
    if len(dst) >0:
        meanDistance = np.mean(dst)
        return 1 - meanDistance
    else:
        return 0 #i.e., one of the lists is empty        
```

```python
def jaccard_similarity(list1, list2):
    # intersection of two sets
    set1 = set(list1)
    set2 = set(list2)
    intersection = len(set1.intersection(set2))
    # Unions of two sets
    union = len(set1.union(set2))
    strRep = str(list(set1.intersection(set2)))
    strRep = strRep.replace("[","") 
    strRep = strRep.replace("]","") 
    strRep = strRep.replace("'","") 

    return (intersection / union, strRep)
```
