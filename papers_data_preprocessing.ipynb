```python
import re

def systemize_groupSizes(gsString):
    gsString = gsString.lower().replace(" ","") 
    if gsString.startswith("avg:"):
        hasAvg = True
    gsList = re.split('\,|\;', gsString)
    parsedList = []
    for f in gsList:
        f = f.replace("avg:","").replace("avg.","").replace("avg","")  
        f = re.sub("[^0-9\.\-]", "", f)
        try:
            if "-" in f: # range value, use just average
                fList = f.split("-")
                fList = [float(i) for i in fList]
                f = np.round(np.mean(fList))

            f = np.round(float(f))
            if f >0:
                #print(f)
                parsedList.append(f)
        except:
            print("error in "+gsString, gsList, f)
    return parsedList            
```

```python
#for each of reported values, get its best fit; then report on the average fit for list 1 
# (i.e., to what extent are sizes used in list1 reflected in list2)
def groupSizeSimilarity(list1, list2):
    # intersection of two sets
    dst = []
    for i in list1:
        dst_i = []
        for j in list2:
            d = abs(i-j) / np.max([i,j])
            dst_i.append(d)
        if len(dst_i) >0:
            dst.append(np.min(dst_i))
    if len(dst) >0:
        meanDistance = np.mean(dst)
        return 1 - meanDistance
    else:
        return 0 #i.e., one of the lists is empty        
```

```python
def jaccard_similarity(list1, list2):
    # intersection of two sets
    set1 = set(list1)
    set2 = set(list2)
    intersection = len(set1.intersection(set2))
    # Unions of two sets
    union = len(set1.union(set2))
    strRep = str(list(set1.intersection(set2)))
    strRep = strRep.replace("[","") 
    strRep = strRep.replace("]","") 
    strRep = strRep.replace("'","") 

    return (intersection / union, strRep)
```

```python
#item1, item2, edgeType, edgeWeight
edgeListBaselines = []
edgeListDatasets = []
edgeListGroupConstruction = []
edgeListEvalMetrics = []
edgeListOther = []

for i,r1 in raw_data.iterrows():
    for j,r2 in raw_data.iterrows():
        if i < j:
            val, simBaselines = jaccard_similarity(r1["ParsedBaselines"],r2["ParsedBaselines"])
            if val > 0:
                edgeListBaselines.append({"item1":i, "item2":j, "type":"same_baseliens","value":val, "label":simBaselines})
                
            val, simDatasets = jaccard_similarity(r1["ParsedDatasetHighLevel"],r2["ParsedDatasetHighLevel"])
            if val > 0:
                edgeListDatasets.append({"item1":i, "item2":j, "type":"same_datasets","value":val, "label":simDatasets}) 
                
            val, simGCS = jaccard_similarity(r1["ParsedGroupConstructionHighLevel"],r2["ParsedGroupConstructionHighLevel"])
            if val > 0:
                edgeListGroupConstruction.append({"item1":i, "item2":j, "type":"same_groupConstruction","value":val, "label":simGCS}) 

            val, simEvDim = jaccard_similarity(r1["EvaluationDimensionsHighLevel"],r2["EvaluationDimensionsHighLevel"])
            if val > 0:
                edgeListEvalMetrics.append({"item1":i, "item2":j, "type":"similar_EvalMetrics","value":val, "label":simEvDim}) 
                
                
            if r1["StudyType"] == r2["StudyType"]:
                edgeListOther.append({"item1":i, "item2":j, "type":"same_studyType","value":1.0, "label":r1["StudyType"]}) 
            if r1["DataPyramid"] == r2["DataPyramid"]:
                edgeListOther.append({"item1":i, "item2":j, "type":"same_dataPyramid","value":1.0, "label":r1["DataPyramid"]}) 
            if r1["Coupled/Decoupled"] == r2["Coupled/Decoupled"]:
                edgeListOther.append({"item1":i, "item2":j, "type":"same_evalStyle","value":1.0, "label":r1["Coupled/Decoupled"]})                 
            if r1["GroundTruthType"] == r2["GroundTruthType"]:
                edgeListOther.append({"item1":i, "item2":j, "type":"same_groundTruth","value":1.0, "label":r1["GroundTruthType"]})                 
            if r1["RecommendationType"] == r2["RecommendationType"]:
                edgeListOther.append({"item1":i, "item2":j, "type":"same_recsType","value":1.0, "label":r1["RecommendationType"]})                 
            if r1["GroupDuration"] == r2["GroupDuration"]:
                edgeListOther.append({"item1":i, "item2":j, "type":"same_groupDuration","value":1.0, "label":r1["GroupDuration"]})                 
                
                
```
