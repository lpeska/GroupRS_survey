## Construct Edges

```python
def clean_list(listIn):
    blacklist = ["other","item-based","-","Not clear","Unclear / not specified","unknown / not applicable"]
    return [i for i in listIn if i not in blacklist]


def jaccard_similarity(list1, list2):
    # intersection of two sets
    set1 = set(clean_list(list1))
    set2 = set(clean_list(list2))
    intersection = len(set1.intersection(set2))
    # Unions of two sets
    union = len(set1.union(set2))
    strRep = str(list(set1.intersection(set2)))
    try:
        return (intersection / union, strRep)
    except:
        return (0,"")
```

```python
#for each of reported values, get its best fit; then report on the average fit for list 1 
# (i.e., to what extent are sizes used in list1 reflected in list2)
def groupSizeSimilarity(list1, list2):
    # intersection of two sets
    dst = []
    for i in list1:
        dst_i = []
        for j in list2:
            d = abs(i-j) / np.max([i,j])
            dst_i.append(d)
        if len(dst_i) >0:
            dst.append(np.min(dst_i))
    if len(dst) >0:
        meanDistance = np.mean(dst)
        return 1 - meanDistance
    else:
        return 0 #i.e., one of the lists is empty  
```

```python
link_width_factor = 10
```

```python
raw_data = reportDF
#item1, item2, edgeType, edgeWeight
edgeListAuthors = []
edgeListVenue = []
edgeListYear = []

for i,r1 in raw_data.iterrows():
    for j,r2 in raw_data.iterrows():
        if i < j:
            val, simAuthors = jaccard_similarity(r1["bib_authors"],r2["bib_authors"])
            if val > 0:
                 
                edgeListAuthors.append({"source":i, "target":j, "type":"same_authors","value":val*link_width_factor, "label":simAuthors})

            if r1["bib_venue"] == r2["bib_venue"]:
                edgeListVenue.append({"source":i, "target":j, "type":"same_venue","value":1.0*link_width_factor, "label":r1["bib_venue"]}) 
            if r1["bib_year"] == r2["bib_year"]:
                edgeListYear.append({"source":i, "target":j, "type":"same_year","value":1.0*link_width_factor, "label":r1["bib_year"]})                 
```

```python
for i,r1 in raw_data.iterrows():
    for j,r2 in raw_data.iterrows():
        if int(r2["bib_year"]) - int(r1["bib_year"]) >0: #first is older
            val = 1/( 1 + int(r2["bib_year"]) - int(r1["bib_year"]))
            if val > 0.1:
                label = str(r1["bib_year"]) + " vs." + str(r2["bib_year"])
                edgeListYear.append({"source":i, "target":j, "type":"year_olderThan","value":val*link_width_factor, "label":label})
```

```python
nodeList = []
for i,r1 in raw_data.iterrows():
    nodeList.append({"id":i, "label":r1["bib_ID"], "title":r1["bib_title"], "year":r1["bib_year"], "venue":r1["bib_venue"]})
```

```python
import requests
import json
import time
import functools

# For some reason, I found some DOIs which were not found in semanticscholar
# Later I noticed, these are DOIs which even when accessed from doi.org leads to paper
# Which actually has different DOI (seem like DOI was changed or something)
# Here is mapping of such DOIs
doi_mapping = {
    # https://ieeexplore.ieee.org/document/7436655/
    "10.1109/mis.2017.2581329": "10.1109/MIS.2016.28",
    # Pre-print doi
    "10.21203/rs.3.rs-2145572/v1": "10.1007/s11063-023-11376-0",
    "10.1145/3091478.3091485": None
}
def map_doi(paper_doi):
    if paper_doi in doi_mapping:
        return doi_mapping[paper_doi]
    return paper_doi

# Sometimes paper could not be found on semantic scholar as it does not have DOI indexed
# In that case we can use different paper ID
doi_to_paper_id_mapping = {
    "10.1145/3091478.3091485": "7471084462634120dad92e17de9c42f4214c8a75"
}
    
def map_doi_to_id(paper_doi):
    return doi_to_paper_id_mapping[paper_doi]

def build_url(identifier, offset, use_doi):
    if use_doi:
        return f"https://api.semanticscholar.org/graph/v1/paper/DOI:{identifier}/citations?fields=externalIds&offset={offset}&limit=500"
    else:
        return f"https://api.semanticscholar.org/graph/v1/paper/{identifier}/citations?fields=externalIds&offset={offset}&limit=500"

bib_id_to_paper_id = {
    "PeskaM21": "380779d60aab8593420e72a2c81436ce0db6c48c",
    "7c9bd078ab504f9ba94dcc4322f8acfa": "9b834140890245e73a32cb74412ef84779173536"
}

# Unknown / missing
# DOI:10.1016/j.eswa.2017.10.027
# The paper cannot be find at semanticscholar at all, it seem to be very hard to access without institutional login

    
# Given paper DOI at the input, returns DOIs of all papers that cite the given paper_doi
@functools.cache
def fetch_cited_by(paper_doi, bib_id):
    paper_doi_orig = paper_doi
    
    if not paper_doi:
        # If it is empty/None at the input, we have no DOI at all, so we take paper id 
        url = build_url(bib_id_to_paper_id[bib_id], 0, False)
    else:
        paper_doi = map_doi(paper_doi)
    
    
        if paper_doi is None:
            url = build_url(map_doi_to_id(paper_doi_orig), 0, False)
        else:
            url = build_url(paper_doi, 0, True)

    dat = []
    offset = 0
    res = requests.get(url)
    
    if res.status_code != 200:
        print(f"Returned: {res.status_code}, {res.text}, {url}")
        return dat

    d = json.loads(res.content)
    dat = dat + [x['citingPaper']['externalIds']['DOI'] for x in d['data'] if 'DOI' in x['citingPaper']['externalIds']]
    
    while 'next' in d:
        assert d['next'] > offset
        offset = d['next']
        time.sleep(2)
        if paper_doi is None:
            if not paper_doi_orig:
                url = build_url(bib_id_to_paper_id[bib_id], offset, False)
            else:
                url = build_url(map_doi_to_id(paper_doi), offset, False)
        else:
            url = build_url(paper_doi, offset, True)
        res = requests.get(url)
        if res.status_code != 200:
            print(f"Returned: {res.status_code}, {res.text}, {url}")
            break
        d = json.loads(res.content)
        dat = dat + [x['citingPaper']['externalIds']['DOI'] for x in d['data'] if 'DOI' in x['citingPaper']['externalIds']]
        
    
    return dat
```

```python
#item1, item2, edgeType, edgeWeight
edgeListBaselines = []
edgeListDatasets = []
edgeListGroupConstruction = []
edgeListEvalMetrics = []
edgeListOther = []
edgeListIsBaselineFor = []
edgeListIsCitedBy = []

for i,r1 in raw_data.iterrows():
    for j,r2 in raw_data.iterrows():
        if i < j:
            val, simBaselines = jaccard_similarity(r1["ParsedBaselines"],r2["ParsedBaselines"])
            if val > 0:
                edgeListBaselines.append({"source":i, "target":j, "type":"same_baseliens","value":val*link_width_factor, "label":simBaselines})
                
            val, simDatasets = jaccard_similarity(r1["ParsedDatasetHighLevel"],r2["ParsedDatasetHighLevel"])
            if val > 0:
                edgeListDatasets.append({"source":i, "target":j, "type":"same_datasets","value":val*link_width_factor, "label":simDatasets}) 
                
            val, simGCS = jaccard_similarity(r1["ParsedGroupConstructionHighLevel"],r2["ParsedGroupConstructionHighLevel"])
            if val > 0:
                edgeListGroupConstruction.append({"source":i, "target":j, "type":"same_groupConstruction","value":val*link_width_factor, "label":simGCS}) 

            val, simEvDim = jaccard_similarity(r1["EvaluationDimensionsHighLevel"],r2["EvaluationDimensionsHighLevel"])
            if val > 0:
                edgeListEvalMetrics.append({"source":i, "target":j, "type":"similar_EvalMetrics","value":val*link_width_factor, "label":simEvDim}) 
                
                
            if r1["StudyType"] == r2["StudyType"]:
                edgeListOther.append({"source":i, "target":j, "type":"same_studyType","value":1.0*link_width_factor, "label":r1["StudyType"]}) 
            if r1["DataPyramid"] == r2["DataPyramid"]:
                edgeListOther.append({"source":i, "target":j, "type":"same_dataPyramid","value":1.0*link_width_factor, "label":r1["DataPyramid"]}) 
            if r1["Coupled/Decoupled"] == r2["Coupled/Decoupled"]:
                edgeListOther.append({"source":i, "target":j, "type":"same_evalStyle","value":1.0*link_width_factor, "label":r1["Coupled/Decoupled"]})                 
            if r1["GroundTruthType"] == r2["GroundTruthType"]:
                edgeListOther.append({"source":i, "target":j, "type":"same_groundTruth","value":1.0*link_width_factor, "label":r1["GroundTruthType"]})                 
            if r1["RecommendationType"] == r2["RecommendationType"]:
                edgeListOther.append({"source":i, "target":j, "type":"same_recsType","value":1.0*link_width_factor, "label":r1["RecommendationType"]})                 
            if r1["GroupDuration"] == r2["GroupDuration"]:
                edgeListOther.append({"source":i, "target":j, "type":"same_groupDuration","value":1.0*link_width_factor, "label":r1["GroupDuration"]})                 
                
            # Build is_baseline_for
            for b in get_baselines(r1.Baselines).keys():
                if b in baseline_name_to_paper_url and baseline_name_to_paper_url[b] == r2.Link:
                    # R2 is baseline for R1
                    edgeListIsBaselineFor.append({"source":j, "target": i, "type": "is_baseline_for", "label": "is_baseline_for", "value": 1.0*link_width_factor})
            
            for b in get_baselines(r2.Baselines).keys():
                if b in baseline_name_to_paper_url and baseline_name_to_paper_url[b] == r1.Link:
                    # R1 is baseline for R2
                    edgeListIsBaselineFor.append({"source":i, "target": j, "type": "is_baseline_for", "label": "is_baseline_for", "value": 1.0*link_width_factor})
                    
                    
            # Build is_cited_by [r1 is cited by r2]
            r1_citations = fetch_cited_by(r1.bib_doi, r1.bib_ID)
            for cit_doi in r1_citations:
                # Just to be sure that there are no minor differences (additional /, spaces etc)
                # That is why we do it like this instead of ==
                if r2.bib_doi in cit_doi or cit_doi in r2.bib_doi:
                    edgeListIsCitedBy.append({"source":i, "target": j, "type": "is_cited_by", "label": "is_cited_by", "value": 1.0*link_width_factor})
                    
            # [r2 is cited by r1]
            r2_citations = fetch_cited_by(r2.bib_doi, r2.bib_ID)
            for cit_doi in r2_citations:
                if r1.bib_doi in cit_doi or cit_doi in r1.bib_doi:
                    edgeListIsCitedBy.append({"source":j, "target": i, "type": "is_cited_by", "label": "is_cited_by", "value": 1.0*link_width_factor})
                    
            print(f"Processed: {i},{j}")
```
